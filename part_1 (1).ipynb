```python
import csv
import json
with open("train.csv", "r", encoding="utf-8") as csv_file:
    reader = csv.DictReader(csv_file)
    data = list(reader)
with open("titanic.json", "w", encoding="utf-8") as json_file:
    json.dump(data, json_file, indent=4)

print("CSV file has been converted to titanic.json successfully!")
```
