```python
with open("train.csv", "r") as file:
    lines = file.readlines()

with open("survivors.csv", "w") as new_file:
    # Write the header
    new_file.write(lines[0])

    count = 0

    # Check each row
    for line in lines[1:]:
        columns = line.strip().split(",")

        if columns[1] == "1":
            count += 1
            new_file.write(line)

print("Number of survivors:", count)
```
